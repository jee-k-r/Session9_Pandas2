```python
import pandas as pd
```

```python
#read csv 
baby_names = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv')
baby_names.info()
baby_names.head()
```

```python
#1. Delete unnamed columns
del baby_names['Unnamed: 0']
baby_names.head()
```

```python
#2. Show the distribution of male and female
baby_names['Gender'].value_counts()
```

```python
#3. Show the top 5 most preferred names
# you don't want to sum the Year column, so you delete it
del baby_names["Year"]

# group the data
names = baby_names.groupby("Name").sum()

# print the first 5 observations
names.head()

# print the size of the dataset
print(names.shape)

# sort it from the biggest value to the smallest one
names.sort_values("Count", ascending = 0).head()
```

```python
#What is the median name occurence in the dataset
names[names.Count == names.Count.median()]
```

```python
#Distribution of male and female born count by states
# you don't want to sum the Id column, so you delete it
del baby_names["Id"]
gender_state_sum = baby_names.groupby(['Gender','State']).sum()
```
